```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
kappa = 1
t0 = 0.5

a = lambda k: 1/np.sqrt(2*np.log(np.log(k)))
b = lambda k: 1/a(k) + 0.5*a(k)*np.log(np.log(np.log(k)))


def beta(n):
    return (np.sqrt(n)*(1 - int(n*t0)/n)*t0 )/ np.sqrt(t0*(1-t0))
def c(alpha, n):
    return b(n) - a(n)*np.log(0.5*np.sqrt(np.pi)*np.log(1/(1-alpha)))
    
def u(alpha, n):
    return (c(alpha,n) + norm.ppf((1- alpha)**(1/(2*np.log(0.5*n)))))**2

def prob(kappa, alpha, n):
    return 1 - norm.cdf(beta(n)*kappa + u(alpha,n)) + norm.cdf(beta(n)*kappa - u(alpha,n))
```

```python
n = int(1000)
alpha = 0.001
d = 1

kappas = np.linspace(0,1e1, 100)
probs = [prob(k, alpha/d, n) for k in kappas]
plt.plot(
    kappas,
    probs
)
```
